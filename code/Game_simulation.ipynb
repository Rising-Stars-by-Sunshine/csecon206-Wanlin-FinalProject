## Scenario 1: The players perfectly meet each other's ideal standards.



In this scenario, payoff matrix is listed as follows:
image.png

```python
import nashpy as nash
import numpy as np

# Creater the game with the payoff matrix

A = np.array([[10, -10,4],
              [-10, -5,-2],
              [3,-3,5]]) # A is the row player (in this case, prisoner)

B = np.array([[10, -10,3],
              [-10, -5,-3],
              [4,-2,5]]) # B is the column player
```

```python
# Form the game
game1 = nash.Game(A,B)
game1
```

```python
# Find the Nash Equilibrium with Support Enumeration
equilibria = game1.support_enumeration()
for eq in equilibria:
    print(eq)
```

**Analysis**

Game 1 has three Nash equilibria, one appearing on “date/date,” one appearing on “keep courting/ keep courting,” and one being mix strategy with a probability of 0.125 to date and 0.875 to keep courting.

These Nash equilibria suggest that neither player has the incentive to choose "Not to Date." Their preferred outcomes are "Date/Date" or "Keep courting/Keep courting." The result from the mixed strategy indicates a higher probability of choosing "Keep courting" between the options of "Date" and "Keep courting."

## Scenario 2: The players partially meet each other's ideal standards, and both are altruistic.



The payoff matrix is as follows:
image.png

```python
import nashpy as nash
import numpy as np

# Creater the game with the payoff matrix

A = np.array([[2, 0,1],
              [-2, -1,-1],
              [1,0,5]]) # A is the row player (in this case, prisoner)

B = np.array([[2, -2,1],
              [0, -1,0],
              [1,-1,5]]) # B is the column player
```

```python
# Form the game
game2 = nash.Game(A,B)
game2
```

```python
# Find the Nash Equilibrium with Support Enumeration
equilibria = game2.support_enumeration()
for eq in equilibria:
    print(eq)
```

**Analysis**

Game 2 has three Nash equilibria, one appearing on “Date/Date,” one appearing on “Keep courting/ Keep courting,” and one being mix Strategy with a probability of 0.8 for date and 0.2 for keep courting.

These Nash equilibria suggest that their preferred outcome is either "Date/Date" or "Keep courting/Keep courting". The result from the mixed strategy indicates a higher probability for the player to choose "Date" between the options of "Date" and "Keep courting."

## Scenario 3: The players partially meet each other's ideal standards, and both are egoistic.

The payoff matrix is listed as follows:

```python
# Creater the game with the payoff matrix

A = np.array([[2, -1,-2],
              [-1, 5,-1],
              [2,-1,-3]]) # A is the row player (in this case, prisoner)

B = np.array([[2, -1,2],
              [-1, 5,-1],
              [-2,-1,-3]]) # B is the column player
```

```python
# Form the game
game3 = nash.Game(A,B)
game3
```

```python
# Find the Nash Equilibrium with Support Enumeration
equilibria = game3.support_enumeration()
for eq in equilibria:
    print(eq)
```

**Analysis**

Game3 has three Nash equilibria, one appearing on “date/date,” one appearing on “not to date/not to date,” and one being mix strategy, with probabilities of 0.6666667 to date and 0.33333333 not to date.

These Nash equilibria suggest that their preferred outcome is "Date/Date" or "Not to Date/Not to Date." The result from the mixed strategy indicates a higher probability of the player choosing "Date" between the options of "Date" and "Not to Date."
